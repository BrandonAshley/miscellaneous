```python
bins = [18, 28, 38, 48, 58, 68, 150]
labels = [1, 2, 3, 4, 5, 6]

# Create a new column 'AgeBin' in the DataFrame
df['AgeBin'] = pd.cut(df['age'], bins=bins, labels=labels)

# Calculate the mean value for each column (except 'id', 'smoking', and 'age') for each age bin
columns_to_exclude = ['id', 'smoking', 'age','AgeBin']
columns_to_process = [col for col in df.columns if col not in columns_to_exclude]

for col in columns_to_process:
    mean_values = df.groupby('AgeBin')[col].transform('mean')
    df[f'{col}Difference'] = df[col] - mean_values

# Drop the 'AgeBin' column
df.drop(columns=['AgeBin'], inplace=True)

# Assuming you also have a 'test_data' DataFrame:

# Create the 'AgeBin' column in the test_data DataFrame
test_data['AgeBin'] = pd.cut(test_data['age'], bins=bins, labels=labels)

# Calculate the mean value for each column (except 'id', 'smoking', and 'age') for each age bin in the test_data DataFrame
for col in columns_to_process:
    mean_values = test_data.groupby('AgeBin')[col].transform('mean')
    test_data[f'{col}Difference'] = test_data[col] - mean_values

# Drop the 'AgeBin' column from the test_data DataFrame
test_data.drop(columns=['AgeBin'], inplace=True)

df.head(10)
```
